```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# CIFAR10 데이터셋 불러오기
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])
trainset = datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
testset = datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)

trainloader = DataLoader(trainset, batch_size=256, shuffle=True)
testloader = DataLoader(testset, batch_size=256, shuffle=False)

# MLP 모델 정의
class MLP(nn.Module):
    def __init__(self, activation_fn):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(3*32*32, 512)
        self.fc2 = nn.Linear(512, 10)
        self.activation = activation_fn

    def forward(self, x):
        x = x.view(-1, 3*32*32)  # CIFAR-10의 입력 크기 (3x32x32)
        x = self.activation(self.fc1(x))
        x = self.fc2(x)
        return x

# 모델 학습 함수
def train_model(model, optimizer, criterion, trainloader, epochs=50):
    train_accuracies = []
    for epoch in range(epochs):
        correct = 0
        total = 0
        model.train()  # 모델 학습 모드로 설정
        for inputs, labels in trainloader:
            inputs, labels = inputs.cuda(), labels.cuda()
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        train_accuracy = 100 * correct / total
        train_accuracies.append(train_accuracy)
    return train_accuracies

# Adam 옵티마이저를 사용할 경우
model_adam = MLP(nn.ReLU()).cuda()
optimizer_adam = optim.Adam(model_adam.parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss()

train_acc_adam = train_model(model_adam, optimizer_adam, criterion, trainloader)

# SGD 옵티마이저를 사용할 경우
model_sgd = MLP(nn.ReLU()).cuda()
optimizer_sgd = optim.SGD(model_sgd.parameters(), lr=0.01, momentum=0.9)

train_acc_sgd = train_model(model_sgd, optimizer_sgd, criterion, trainloader)

# Plot: Adam과 SGD 학습 정확도 비교
plt.figure(figsize=(10, 5))
plt.plot(train_acc_adam, label="Adam")
plt.plot(train_acc_sgd, label="SGD")
plt.xlabel("Epochs")
plt.ylabel("Train Accuracy (%)")
plt.legend()
plt.title("SGD vs Adam: Train Accuracy")
plt.show()
```

```python
# 모델 정의 (LeakyReLU와 Sigmoid 사용)
class MLP_Sigmoid(nn.Module):
    def __init__(self):
        super(MLP_Sigmoid, self).__init__()
        self.fc1 = nn.Linear(3*32*32, 512)
        self.fc2 = nn.Linear(512, 10)
        self.activation = nn.Sigmoid()

    def forward(self, x):
        x = x.view(-1, 3*32*32)  # CIFAR-10 입력 크기
        x = self.activation(self.fc1(x))
        x = self.fc2(x)
        return x

class MLP_LeakyReLU(nn.Module):
    def __init__(self):
        super(MLP_LeakyReLU, self).__init__()
        self.fc1 = nn.Linear(3*32*32, 512)
        self.fc2 = nn.Linear(512, 10)
        self.activation = nn.LeakyReLU()

    def forward(self, x):
        x = x.view(-1, 3*32*32)  # CIFAR-10 입력 크기
        x = self.activation(self.fc1(x))
        x = self.fc2(x)
        return x

# 모델 학습 함수 (Sigmoid와 LeakyReLU)
model_sigmoid = MLP_Sigmoid().cuda()
optimizer_sigmoid = optim.Adam(model_sigmoid.parameters(), lr=0.001)

model_leakyrelu = MLP_LeakyReLU().cuda()
optimizer_leakyrelu = optim.Adam(model_leakyrelu.parameters(), lr=0.001)

train_acc_sigmoid = train_model(model_sigmoid, optimizer_sigmoid, criterion, trainloader)
train_acc_leakyrelu = train_model(model_leakyrelu, optimizer_leakyrelu, criterion, trainloader)

# Plot: Sigmoid vs LeakyReLU 학습 정확도 비교
plt.figure(figsize=(10, 5))
plt.plot(train_acc_sigmoid, label="Sigmoid")
plt.plot(train_acc_leakyrelu, label="LeakyReLU")
plt.xlabel("Epochs")
plt.ylabel("Train Accuracy (%)")
plt.legend()
plt.title("Sigmoid vs LeakyReLU: Train Accuracy")
plt.show()
```

```python
# Dropout을 적용한 모델 정의
class MLP_Dropout(nn.Module):
    def __init__(self):
        super(MLP_Dropout, self).__init__()
        self.fc1 = nn.Linear(3*32*32, 512)
        self.fc2 = nn.Linear(512, 10)
        self.activation = nn.LeakyReLU()
        self.dropout = nn.Dropout(0.1)  # Dropout 확률 0.1

    def forward(self, x):
        x = x.view(-1, 3*32*32)  # CIFAR-10 입력 크기
        x = self.dropout(self.activation(self.fc1(x)))  # Dropout 적용
        x = self.fc2(x)
        return x

# Dropout 모델 학습
model_dropout = MLP_Dropout().cuda()
optimizer_dropout = optim.Adam(model_dropout.parameters(), lr=0.001)

# 학습 및 정확도 추적
train_acc_dropout = train_model(model_dropout, optimizer_dropout, criterion, trainloader)

# 테스트 정확도 계산
def test_model(model, testloader):
    correct = 0
    total = 0
    model.eval()  # 평가 모드로 설정
    with torch.no_grad():
        for inputs, labels in testloader:
            inputs, labels = inputs.cuda(), labels.cuda()
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    return 100 * correct / total

test_acc_dropout = test_model(model_dropout, testloader)

# Plot: Dropout 적용한 모델 학습 정확도 비교
plt.figure(figsize=(10, 5))
plt.plot(train_acc_dropout, label="Train Accuracy with Dropout")
plt.axhline(y=test_acc_dropout, color='r', linestyle='--', label=f"Test Accuracy: {test_acc_dropout:.2f}%")
plt.xlabel("Epochs")
plt.ylabel("Accuracy (%)")
plt.legend()
plt.title("Dropout Applied: Train vs Test Accuracy")
plt.show()
```
